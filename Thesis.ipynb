## <strong>Noțiuni introductive</strong>
</div>

Documentul pe care îl citiți nu este o pagină web statică, ci un mediu interactiv denumit <strong>blocnotes Colab</strong>, în care puteți să scrieți și să executați cod.

De exemplu, aceasta este o <strong>celulă de cod</strong> cu un script Python scurt care calculează o valoare, o stochează într-o variabilă și printează rezultatul:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Pentru a executa codul din celula de mai sus, selectați celula cu un clic, apoi apăsați butonul Redați din stânga codului sau folosiți comanda rapidă de la tastatură „Command / Ctrl + Enter”. Pentru a edita codul, dați clic pe celulă și faceți modificări.

Variabilele pe care le definiți într-o celulă pot fi folosite ulterior în alte celule:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Cu ajutorul blocnotesurilor Colab, puteți să combinați <strong>cod executabil</strong> și <strong>text îmbogățit</strong> într-un singur document, împreună cu <strong>imagini</strong>, <strong>HTML</strong>, <strong>LaTeX</strong> și altele.
```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import numpy as np
import tensorflow as tf
import pickle
import cv2
import os
# import tensorflow_addons as tfa
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Activation, Flatten, Dropout, Dense, BatchNormalization, GlobalAveragePooling2D
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

# PREPROCESSING

```
learning_rate = 0.0001
batch_size = 32
image_size = tuple((224, 224))
train_dir = './train'
valid_dir = './valid'
width=224
height=224
depth=3

datagen = ImageDataGenerator(
    rescale=1./255,
    vertical_flip=True,
    horizontal_flip=True,
    )
validgen = ImageDataGenerator(
    rescale=1./255,
    vertical_flip=True,
    horizontal_flip=True, validation_split=0.5
    )

train_it = datagen.flow_from_directory(train_dir,
        class_mode='categorical',
        batch_size=32,
        target_size=image_size)
valid_it = validgen.flow_from_directory(valid_dir,
        class_mode='categorical',
        batch_size=32,
        target_size=image_size,
        subset='training')
test_it = validgen.flow_from_directory(valid_dir,
        class_mode='categorical',
        batch_size=32,
        target_size=image_size,
        subset='validation')

target_names=train_it.class_indices
print(target_names);
classes_count = len(target_names)
```

# MODEL IMPLEMENTATION

# Model with Insection V3

```
model = Sequential()
input_shape = (width, height, depth)
base_layer = InceptionV3(weights='imagenet',
                        include_top=False,
                        input_shape=input_shape)

for layer in base_layer.layers:
    layer.trainable = False

for layer in base_layer.layers[-4:]:
    layer.trainable = True

# model.add(Conv2D(512, (3, 3), 1, padding='same', activation='relu'))
# # model.add(BatchNormalization())
# model.add(Dropout(0.2))
# model.add(MaxPooling2D(padding='same'))
model.add(base_layer)
model.add(GlobalAveragePooling2D())
# model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.35))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(50, activation='relu'))
model.add(Dense(classes_count, activation='softmax'))
```

# Simple Model

```
model = Sequential()
inputShape = (width, height, depth)
model.add(Conv2D(32, (3, 3), 1, activation='relu', input_shape=inputShape))
# model.add(BatchNormalization())
model.add(MaxPooling2D())

model.add(Conv2D(64, (3, 3), 1, activation='relu'))
# model.add(BatchNormalization())
model.add(Dropout(0.35))
model.add(MaxPooling2D())

# model.add(Conv2D(64, (3, 3), 1, activation='relu'))
# # model.add(BatchNormalization())
# model.add(Dropout(0.3))
# model.add(MaxPooling2D())

model.add(Conv2D(128, (3, 3), 1, activation='relu'))
# model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(MaxPooling2D())

# model.add(GlobalAveragePooling2D())
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(50, activation='relu'))
model.add(Dense(classes_count, activation='softmax'))
```

```
model = Sequential()
inputShape = (width, height, depth)
chanDim = -1
model.add(Conv2D(32, (3, 3), input_shape=inputShape))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))
model.add(MaxPooling2D(pool_size=(3, 3)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))

model.add(Conv2D(64, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))

model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))

model.add(Conv2D(128, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.35))
model.add(Dense(50, activation='relu'))
model.add(Dense(classes_count, activation='softmax'))
```

```
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=[
    'accuracy',
    tf.keras.metrics.Recall(),
    tf.keras.metrics.Precision()])

model.summary()
# train the network
print("[INFO] training network...")
```

```
# logdir = '../../drive/MyDrive/AI/logs'

# tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
EPOCHS = 20
epochsteps = len(train_it)
valsteps = len(valid_it)
teststeps = len(test_it)
history = model.fit(train_it, steps_per_epoch=epochsteps, epochs=EPOCHS, validation_data=valid_it, validation_steps=valsteps, verbose=1,
                    # callbacks=[tensorboard_callback ]
                    )

print("EVALUATION...")
values = model.evaluate(test_it, steps=teststeps, verbose=1)
print(values)
# print('accuracy',acc * 100.0)


print_file = '../../drive/MyDrive/AI/models/4june_inception_unfrozen_diseases/'

for key in history.history.keys():
  file_name = print_file + key + '.txt'
  with open(file_name, 'w') as f:
    for val in history.history[key]:
      print(val, file=f)
with open(print_file + 'evaluation.txt', 'w') as f:
  for val in values:
    print(val, file=f)

model.save('../../drive/MyDrive/AI/models/4june_inception_unfrozen_diseases/model.h5')
```

```
from sklearn.metrics import plot_confusion_matrix

class estimator:
  _estimator_type = ''
  classes_=[]
  def __init__(self, model, classes):
    self.model = model
    self._estimator_type = 'classifier'
    self.classes_ = classes
  def predict(self, X):
    y_prob= self.model.predict(X)
    y_pred = y_prob.argmax(axis=1)
    return y_pred

classifier = estimator(model, list(target_names.keys()))
plot_confusion_matrix(estimator=classifier, X=test_it, y_true=list(target_names.keys()))

```
